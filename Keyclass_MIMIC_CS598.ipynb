```python
import os
import json
import pandas as pd

folder_path = "/content/drive/MyDrive/CS598DLH_KeyClass_Reproduce/Final_Project/keyclass/results/mimic/"

# These arrays will store the respective metrics generated from the experiment
self_train_f1 = []
self_train_prec = []
self_train_rec = []

downstream_f1 = []
downstream_prec = []
downstream_rec = []

# Converting the txt files outputted by the main code into json
# because json is easier to parse in our situation than txt
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        base = os.path.splitext(filename)[0]
        os.rename(folder_path+filename, folder_path+base + '.json')

# Formats the mean and std dev into the format that will eventually be printed below
# It is of the format of "mean +- std dev"
def strformat(val):
    m,s = val
    return str("{:.3f}".format(m))+u"\u00B1"+str("{:.3f}".format(s))

# Takes each logged file containing F1, Precision and Recall from the results folder
# and adds each respective one to the arrays created above
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        with open(os.path.join(folder_path, filename), "r") as f:
            data = json.load(f)
            if "self_trained" in filename:
                self_train_f1.append(strformat(data["F1_Default (mean, std)"]))
                self_train_prec.append(strformat(data["Precision_Default (mean, std)"]))
                self_train_rec.append(strformat(data["Recall_Default (mean, std)"]))
            elif "test_end" in filename:
                downstream_f1.append(strformat(data["F1_Default (mean, std)"]))
                downstream_prec.append(strformat(data["Precision_Default (mean, std)"]))
                downstream_rec.append(strformat(data["Recall_Default (mean, std)"]))


# Converting the formatted metrics into a dataframe
# This is the ablation part of the experiment
self_train = pd.DataFrame()
self_train['F1'] = self_train_f1
self_train['Precision'] = self_train_prec
self_train['Recall'] = self_train_rec
self_train.to_csv('self_train.csv')

# Converting the formatted metrics into a dataframe
# This is from the downstream classifier post refinement
downstream = pd.DataFrame()
downstream['F1'] = downstream_f1
downstream['Precision'] = downstream_prec
downstream['Recall'] = downstream_rec
downstream.to_csv('downstream.csv')
```

```python
# This function takes the values from the metrics two dataframes created above
# and splits them into the mean and std dev
# and then returns the average of the means
def getavg(col):
  l = len(col)
  m = 0
  for r in col:
    splitstr = r.split(u"\u00B1")
    m += float(splitstr[0])
  return m/l

print("Pre Self Train Results:")
print("Average F1 = ",getavg(self_train['F1']))
print("Average Precision = ", getavg(self_train['Precision']))
print("Average Recall = ", getavg(self_train['Recall']))
print()
print("Downstream Classifier Results:")
print("Average F1 = ", getavg(downstream['F1']))
print("Average Precision = ", getavg(downstream['Precision']))
print("Average Recall = ", getavg(downstream['Recall']))
```
